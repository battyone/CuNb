```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# Load data
def load_data(filename):
    directory = './'
    data = np.genfromtxt(directory+filename, skip_header=35, delimiter=',')
    deux_theta = data[:, 0]
    I = data[:, 1]
    return deux_theta, I
```

```python
deuxtheta, I = load_data('2019_theta_2theta_LaB6_detecteurFente025.csv')

plt.plot(deuxtheta, I);
```

```python
def crop(X, Y, x0, x_range):
    mask = np.logical_and(X > x0-x_range/2, X < x0+x_range/2)
    return X[mask], Y[mask]
```

```python
plt.plot(*crop(deuxtheta, I, 53.885, 2));
```

```python
x, y = crop(deuxtheta, I, 54, 2)
```
