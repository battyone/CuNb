```python
import numpy as np
import matplotlib.pylab as plt
```

```python
lmbda = 0.25
a = 1
```

```python
def T(ui, ui_moins1):
    ui_plus1 = -ui_moins1 + 2*ui + lmbda*np.pi/2/a*np.sin(np.pi/a*ui)
    
    return [ui_plus1, ui]

def Ts(p_i, theta_i):
    p_iPlus1 = p_i + lmbda*np.pi/2/a*np.sin(np.pi/a*theta_i)
    theta_iPlus1 = ( p_iPlus1 + theta_i ) % (2*a)
    return [p_iPlus1, theta_iPlus1]
```

```python
def iterate(Pzero, n=100):
    P = [Pzero]

    for k in range(n):
        P.append(Ts(*P[-1]))

    return list( zip(*P) )
```

```python
plt.figure(figsize=(14, 8))
plt.plot(*iterate([0.2, .2], n=8000), ',r')

plt.plot(*iterate([0.01, .001], n=8000), ',b')

plt.plot(*iterate([0.0, 1.1], n=5000), ',m')
plt.plot(*iterate([0.0, 1.2], n=5000), ',m')
plt.plot(*iterate([0.0, 1.3], n=5000), ',m')
plt.plot(*iterate([0.0, 1.4], n=5000), ',m')
plt.plot(*iterate([0.0, 1.5], n=5000), ',m')
plt.plot(*iterate([0.0, 1.55], n=5000), ',k')

plt.plot(*iterate([0.52, 0.30], n=17000), ',g')

plt.axis('equal'); plt.ylabel('theta_i'); plt.xlabel('p_i');
plt.ylim((0, 2)); plt.xlim((-1.5, 2.0));
```
